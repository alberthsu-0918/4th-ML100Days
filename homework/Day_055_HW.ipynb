- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets
import matplotlib
import matplotlib.pyplot as plt

iris = datasets.load_iris()
#X = iris.data
X = iris.data[:, :2]  # we only take the first two features.
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
from sklearn.cluster import KMeans
import numpy as np

#For two clusters
estimators =  KMeans(n_clusters=2)
estimators.fit(X)
labels = estimators.labels_
```

```python
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
plt.figure(2, figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels.astype(np.float), edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
#plt.xticks(())
#plt.yticks(())
plt.show()
```

```python
# for three clusters
estimators =  KMeans(n_clusters=3)
estimators.fit(X)
labels = estimators.labels_
plt.figure(2, figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels.astype(np.float), edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
#plt.xticks(())
#plt.yticks(())
plt.show()
```

```python
# try 3 clusters with other init
estimators =  KMeans(n_clusters=3, n_init=2, init='random')
estimators.fit(X)
labels = estimators.labels_
plt.figure(2, figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels.astype(np.float), edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
#plt.xticks(())
#plt.yticks(())
plt.show()
```
