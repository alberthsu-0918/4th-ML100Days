### 用 digits (3個數字即可 , dataset.load_digits(n_class=3)) 資料嘗試跑 PCA


```python
from sklearn import datasets
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#digits = datasets.load_digits(n_class=3)
digits = datasets.load_digits()
X = digits.data
y = digits.target
```

載入 相關套件 並 執行 PCA 實驗 ...

```python
#my goal is to distinguish number 0-9, the hand written data in datasets.load_digits
from sklearn import decomposition

pca = decomposition.PCA(n_components=2)
pca.fit(X)
X_rd = pca.transform(X)

plt.scatter(X_rd[:, 0], X_rd[:, 1], c=y, cmap="Paired")
plt.colorbar()
```

```python
#dimension reduce, 原本抹個data為8x8 (64)個feature，現在壓縮成2d(2 feature)來表示。尚稱得上可辨識。
```
